```python
# Scatter plots for numerical columns
numerical_cols = data2.select_dtypes(include=[np.number]).columns
target_col = 'average price '  # Assuming this is the correct column name

num_cols = len(numerical_cols)
cols_per_row = 3  # Number of columns per row
rows = (num_cols - 1) // cols_per_row + 1  # Calculate number of rows needed

plt.figure(figsize=(16, 12))
for i, col in enumerate(numerical_cols):
    if col != target_col:
        plt.subplot(rows, cols_per_row, i + 1)  # Adjust the grid size dynamically
        sns.scatterplot(x=col, y=target_col, hue='booking status', data=data2)
        plt.title(f'Scatter plot of {col} vs {target_col}')
plt.tight_layout()
plt.show()
```

```python
data2['date of reservation'] = pd.to_datetime(data2['date of reservation'], errors='coerce')

data2['Month'] = data2['date of reservation'].dt.month


cancellations_by_month = data2[data2['booking status'] == 'Canceled'].dropna(subset=['Month']) \
                          .groupby('Month').size().reset_index(name='Cancellation Count')


cancellations_by_month['Month'] = cancellations_by_month['Month'].astype(int)

print(cancellations_by_month['Month'].unique())  # Check unique values to ensure they are numeric

custom_palette = {'1': '#ffb6c1', '2': '#0000cd', '3': '#00fa9a', '4': '#8a2be2', '5': '#ff6347',
                  '6': '#ffd700', '7': '#8b4513', '8': '#4169e1', '9': '#2e8b57', '10': '#ff4500',
                  '11': '#dda0dd', '12': '#32cd32'}

# Plot cancellations by month
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Cancellation Count', data=cancellations_by_month, palette=custom_palette)
plt.title('Cancellations by Month')
plt.xlabel('Month')
plt.ylabel('Cancellation Count')
plt.xticks(rotation=45)
plt.show()
```

```python
data2['date of reservation'] = pd.to_datetime(data2['date of reservation'], errors='coerce')

# Extract month from the date
data2['Month'] = data2['date of reservation'].dt.month

# Count cancellations by month for both 'Canceled' and 'Not Canceled'
cancellations_by_month = data2.groupby(['Month', 'booking status']).size().reset_index(name='Count')

# Convert Month to integers (if necessary)
cancellations_by_month['Month'] = cancellations_by_month['Month'].astype(int)

# Define custom color palette
custom_palette = {'Canceled': '#ff00ff', 'Not_Canceled': '#7b68ee'}

# Plot cancellations by month with hue differentiation
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Count', hue='booking status', data=cancellations_by_month, palette=custom_palette)
plt.title('Cancellations vs. Not Canceled by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Booking Status')
plt.show()
```

```python
data2['date of reservation'] = pd.to_datetime(data2['date of reservation'], errors='coerce')

# Extract year from the date
data2['Year'] = data2['date of reservation'].dt.year

# Count cancellations by year for both 'Canceled' and 'Not Canceled'
cancellations_by_year = data2.groupby(['Year', 'booking status']).size().reset_index(name='Count')

# Define custom color palette
custom_palette = {'Canceled': '#000080', 'Not_Canceled': '#7b68ee'}

# Plot cancellations by year with hue differentiation
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Count', hue='booking status', data=cancellations_by_year, palette=custom_palette)
plt.title('Cancellations vs. Not Canceled by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Booking Status')
plt.show()
```
