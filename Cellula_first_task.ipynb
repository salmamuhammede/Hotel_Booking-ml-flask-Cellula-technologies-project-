```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le
```

```python
cor = data.corr()

# Generate a heatmap
sns.heatmap(cor, xticklabels=cor.columns, yticklabels=cor.columns )

# Show the plot
plt.show()
```

```python
count_0_1 = data['booking status'].value_counts()
colors = sns.color_palette("pastel")

plt.figure(figsize=(8, 8))
plt.pie(
    count_0_1,
    labels=['Canceled', 'Not_Canceled'],
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops=dict(edgecolor='black', linewidth=1.5,alpha=0.9)

)

# Add a title
plt.title('Pie Chart of Categories', fontsize=16)

# Ensure the pie chart is a circle
plt.axis('equal')

# Show the plot
plt.show()
```
